```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sample_size = 500
```

```python
sample = np.linspace(0, 2 * np.pi, sample_size, dtype = np.float64)
```

```python
signal = np.sin(sample)
```

```python
noise = np.random.normal(0, 1, sample_size)
```

```python
noisy_signal = signal + noise / 5
```

```python
plt.plot(sample, noisy_signal, 'b.', alpha = 0.5)
plt.plot(sample, signal, 'r-', linewidth = 4)
plt.legend(['Sensor readings', 'Truth'])
plt.xlabel("Time")
plt.ylabel("Value of the thing we care about")
plt.show()
```

```python
del signal
```

```python
from statsmodels.nonparametric.smoothers_lowess import lowess
```

```python
cleaned = lowess(noisy_signal, sample, frac=0.1)
```

```python
plt.plot(sample, noisy_signal, 'b.', alpha = 0.5)
plt.plot(cleaned[:, 0], cleaned[:, 1], 'r-', linewidth = 4)
plt.legend(['Sensor readings', 'Reconstructed signal'])
plt.show()
```
